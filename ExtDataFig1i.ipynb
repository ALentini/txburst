```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
c57_UMI = pd.read_csv('data/SS3_c57_UMIs_concat.csv', index_col=0)
cast_UMI = pd.read_csv('data/SS3_cast_UMIs_concat.csv', index_col=0)
c57_UMI = c57_UMI.dropna(how='all')
cast_UMI = cast_UMI.dropna(how='all')
c57_UMI = c57_UMI.fillna(0)
cast_UMI = cast_UMI.fillna(0)
```

```python
silent = (c57_UMI == 0)&(cast_UMI == 0)
```

```python
biallelic = (c57_UMI > 0)&(cast_UMI > 0)
```

```python
len(silent)
```

```python
plt.scatter(100*silent.mean(axis=1), 100*biallelic.mean(axis=1), s=1)
plt.xlabel('Silent cells (%)')
plt.ylabel('Biallelic cells (%)')
plt.savefig('figures/ExtDataFig1i.pdf')
plt.show()
```
